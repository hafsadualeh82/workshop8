```python
import networkx as nx

# Create directed graph
G = nx.DiGraph()

# Add edges from the graph
edges = [
    ('A', 'B'),
    ('B', 'C'),
    ('C', 'E'),
    ('E', 'D'),
    ('D', 'B'),
    ('E', 'F')
]

G.add_edges_from(edges)

# Number of nodes and edges
print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())

# Average in-degree
avg_in_degree = sum(dict(G.in_degree()).values()) / G.number_of_nodes()
print("Average in-degree:", avg_in_degree)

# Average out-degree
avg_out_degree = sum(dict(G.out_degree()).values()) / G.number_of_nodes()
print("Average out-degree:", avg_out_degree)

# Degree centrality
print("\nDegree Centrality:")
for node, value in nx.degree_centrality(G).items():
    print(node, ":", round(value, 4))

# Betweenness centrality
print("\nBetweenness Centrality:")
for node, value in nx.betweenness_centrality(G).items():
    print(node, ":", round(value, 4))

# Closeness centrality
print("\nCloseness Centrality:")
for node, value in nx.closeness_centrality(G).items():
    print(node, ":", round(value, 4))
```
